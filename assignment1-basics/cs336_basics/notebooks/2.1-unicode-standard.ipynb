**Problem (unicode1):** Understanding Unicode
- (a) What does `chr(0)` return?
- (b) How does `__repr__()` differ from printed representation?
- (c) What happens when this character occurs in text?

```python
# (a): empty string
print("aaaa"+chr(0)+"bbbb")
```

```python
# (b): a complete sentence for dev
s = "Hello\nworld!"
print(s)
print(repr(s))


s = "Hello\u0000world!"
print(s)
print(repr(s))

```

```python
print(repr(chr(0)))

print(chr(0))

print(repr("this is a test" + chr(0) + "string"))
print("this is a test" + chr(0) + "string")
```

## My Answers

**(a):** `chr(0)` returns the NULL character (`\x00`), an empty/invisible string character.

**(b):** `print()` renders the character visually (NULL shows as space/nothing), while `repr()` shows the escape sequence (`\x00`), useful for debugging.

**(c):** The NULL character is invisible in printed output but present in the string â€” it can cause issues in C-style string handling where `\0` is a terminator, but Python strings handle it fine since they're length-prefixed.
